```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

### Libraries:

1. **`numpy`**: A library used for numerical operations, particularly for handling arrays and matrices. It provides support for large, multi-dimensional arrays and matrices, along with a collection of mathematical functions to operate on these arrays.

2. **`matplotlib.pyplot`**: A plotting library that allows us to visualize data in various formats (e.g., scatter plots, line charts, etc.). In this case, we use it to create scatter plots to visualize the data and the results of K-Means clustering.

3. **`KMeans from sklearn.cluster`**: This is a clustering algorithm provided by the scikit-learn library. K-Means is an unsupervised learning algorithm that groups data points into clusters based on their similarity. It assigns each data point to the cluster whose center (centroid) is closest. It is widely used for clustering tasks in machine learning.

```python
np.random.seed(42)
X = np.random.rand(300, 2) * 10
```

1. **`np.random.seed(42)`**:
   - This sets a random seed to ensure the reproducibility of results. By using the same seed (`42`), you guarantee that running the code multiple times will generate the same random numbers, which helps in comparing results across different runs.

2. **`np.random.rand(300, 2)`**:
   - Generates 300 random points, each having 2 features (or dimensions). The values of these points are between 0 and 1, as this function generates uniformly distributed random numbers in this range.

3. **`* 10`**:
   - Scales the random values generated in the previous step. This ensures that the generated points fall in the range `[0, 10]` instead of `[0, 1]`. This scaling helps in adjusting the scale of the data.

So, the variable `X` is a **2D array** where each row represents a point in a 2-dimensional space, and there are 300 such points. Each of these points has two feature values (coordinates) in the range [0, 10].

```python
k = 4  # Number of clusters
kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
kmeans.fit(X)
```

1. **`k = 4`**:
   - This specifies that we want to divide the data into 4 clusters. The number of clusters (`k`) is a key parameter in the K-Means algorithm, determining how many groups the algorithm will create from the data.

2. **`KMeans(n_clusters=k, random_state=42, n_init=10)`**:
   - `n_clusters=k`: Specifies the number of clusters, which is set to 4 in this case.
   - `random_state=42`: Ensures reproducibility of results by setting a fixed random seed. This ensures that every time the code is run, the same random initialization of centroids happens.
   - `n_init=10`: The algorithm will run 10 times with different centroid seeds. After these 10 runs, the result with the lowest inertia (the sum of squared distances between data points and their centroids) will be selected as the final clustering result.

3. **`kmeans.fit(X)`**:
   - This fits the K-Means model to the data `X`. The `fit()` method finds the optimal centroids that best divide the data into 4 clusters by minimizing the inertia. The algorithm will try to group the 300 data points into 4 clusters based on their similarity.


```python
labels = kmeans.labels_
centroids = kmeans.cluster_centers_
```

1. **`labels = kmeans.labels_`**:
   - After fitting the K-Means model, `kmeans.labels_` contains the cluster assignments for each data point.
   - Each data point is assigned to one of the clusters. For example:
     - If a data point belongs to cluster 0, its label will be `0`.
     - If it belongs to cluster 1, its label will be `1`, and so on.
   - This array allows us to know which data points belong to which cluster after the algorithm has completed.

2. **`centroids = kmeans.cluster_centers_`**:
   - This provides the coordinates of the centroids (center points) of each cluster.
   - Each centroid is the mean position of all the points assigned to that particular cluster.
   - These centroids represent the "center" of each cluster in the feature space, and the K-Means algorithm optimizes these centroids during the fitting process.

```python
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', alpha=0.6)
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='X', s=200, label='Centroids')
plt.title("K-Means Clustering")
plt.legend()
plt.show()
```

### Explanation of Plotting:

1. **`plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', alpha=0.6)`**:
   - `X[:, 0]` and `X[:, 1]`: These represent the x-coordinates and y-coordinates of the data points in the dataset `X`. In this case, `X` is a 2D array where each row is a data point and the columns represent the coordinates.
   - `c=labels`: The color of each point is determined by its cluster label. Points in the same cluster will have the same color. This is useful for visually distinguishing the clusters.
   - `cmap='viridis'`: Specifies the colormap used to assign colors to different clusters. `'viridis'` is a popular colormap that provides a gradient of colors, making it easy to differentiate the clusters.
   - `alpha=0.6`: This makes the points slightly transparent so that overlapping points are easier to see. The alpha value controls the transparency (with `1` being fully opaque and `0` being fully transparent).

2. **`plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='X', s=200, label='Centroids')`**:
   - `centroids[:, 0]` and `centroids[:, 1]`: These are the x and y coordinates of the centroids of the clusters.
   - `c='red'`: The centroids are colored red to make them easily distinguishable from the data points.
   - `marker='X'`: The centroids are marked with an "X" symbol to clearly identify them in the plot.
   - `s=200`: This sets the size of the markers for the centroids. A larger size makes the centroids more prominent.
   - `label='Centroids'`: Adds a label to the centroids for the legend, so that the legend will display "Centroids" next to the red "X" markers.

3. **`plt.title("K-Means Clustering")`**:
   - This sets the title of the plot to "K-Means Clustering", providing context for what the plot represents.

4. **`plt.legend()`**:
   - This adds a legend to the plot. The legend will display the label "Centroids" next to the red "X" markers, helping to explain what the red markers represent.

5. **`plt.show()`**:
   - This displays the plot, rendering the scatter plot with the clusters and centroids on the screen.
