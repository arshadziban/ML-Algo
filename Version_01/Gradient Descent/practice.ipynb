```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

# Dataset

```python
x = np.array([1,2,3])
y = np.array([2,4,6])
```

*   x represents the independent variable (number of lemonades sold).
*   y represents the dependent variable (money made).

These are the data points we want to fit a linear model to.

```python
x,y
```

# Initial Guess for Parameters

```python
m = 1  # slope
b = 2  # intercept
```

* m is the slope of the line  **y=mx+b**.
* b is the intercept of the line.

These are the initial guesses for the parameters of the linear model.

# Learning Rate and Number of Iterations

```python
learning_rate = 0.01
iterations = 2000
```

* learning_rate (α) controls the step size during each iteration of gradient descent.
* iterations is the number of times the algorithm will update the parameters m and b.

# Number of Data Points

```python
n = len(x)
```

*   n is the number of data points, which is 3 in this case.



# Gradient Descent Loop

```python
# Gradient Descent Loop
for i in range(iterations):
    # Predictions using the current m and b
    # y_pred = mx + b
    y_pred = m * x + b

    # Calculate the cost (Mean Squared Error)
    # J(m, b) = (1/n) * Σ(y_pred - y)^2
    cost = (1/n) * np.sum((y_pred - y) ** 2)

    # Calculate the gradients
    # Gradient of the cost function with respect to m
    # ∂J/∂m = (2/n) * Σ(y_pred - y) * x
    m_gradient = (2/n) * np.sum((y_pred - y) * x)

    # Gradient of the cost function with respect to b
    # ∂J/∂b = (2/n) * Σ(y_pred - y)
    b_gradient = (2/n) * np.sum(y_pred - y)

    # Update m and b
    # m = m - α * ∂J/∂m
    # b = b - α * ∂J/∂b
    m = m - learning_rate * m_gradient
    b = b - learning_rate * b_gradient

    # Print the cost every 100 iterations
    if i % 100 == 0:
        print(f"Iteration {i}: Cost = {cost:.4f}, m = {m:.4f}, b = {b:.4f}")
```

# Final Values of m and b

```python
print(f"\nFinal values: m = {m:.4f}, b = {b:.4f}")
```

# Prediction

```python
# Predict the value for x = 4 using the final m and b
x_new = 4
y_pred_new = m * x_new + b
```

```python
y_pred_new
```

# Plotting the Results

```python
plt.scatter(x, y, color='blue', label='Data points')
plt.plot(x, m * x + b, color='red', label='Best fit line (y = mx + b)')
plt.xlabel('Lemonades Sold (x)')
plt.ylabel('Money Made (y)')
plt.title('Gradient Descent: Lemonade Stand')
plt.legend()
plt.show()
```

```python
# Plotting the data and the line
plt.scatter(x, y, color='blue', label='Data points')  # Original data points
plt.plot(x, m * x + b, color='red', label='Best fit line (y = mx + b)')  # Best-fit line
plt.scatter(x_new, y_pred_new, color='green', label=f'Predicted point (x={x_new}, y={y_pred_new:.4f})')  # Predicted point
plt.xlabel('Lemonades Sold (x)')
plt.ylabel('Money Made (y)')
plt.title('Gradient Descent: Lemonade Stand')
plt.legend()
plt.show()
```
